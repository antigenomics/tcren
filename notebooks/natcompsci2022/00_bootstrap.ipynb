# 00 · Bootstrap data

Reproduction of Karnaukhov et al., *Nat Comput Sci* 2022 (`code_paper/`) using **only the tcren
Python pipeline** — no `mir.jar`. All structures come from the Hugging Face dataset
[`isalgo/tcren_structures`](https://huggingface.co/datasets/isalgo/tcren_structures); vdjdb is
pinned to the paper's `2022-03-30` release. This notebook fetches the **inputs** into the shared
`../data/` dir (`notebooks/data/`; HF structures are gitignored) and stages the 2022 mir/R
**comparison baselines** into this notebook's `data_legacy/`. New results are computed only from
`../data/` + the tcren source code.

```python
# Bootstrap (idempotent: HF snapshot + vdjdb + paper data + legacy results)
from tcren.paper import bootstrap
summary = bootstrap()
summary
```

```python
# Inventory
from pathlib import Path
data = Path('../data')  # shared input dir (notebooks/data); comparison baselines in data_legacy/
for folder in ['Native2022', 'PolyV2022', 'Bobisse', 'Bigot']:
    n = len(list((data / 'structures' / folder).glob('*.pdb')))
    print(f'structures/{folder}: {n} pdb')
print('vdjdb:', [p.name for p in (data / 'vdjdb').glob('*.gz')])
print('external inputs (gz):', len(list(data.rglob('*.gz'))))
print('legacy baselines (gz):', len(list(Path('data_legacy').rglob('*.gz'))))
```
